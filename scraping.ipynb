```python
from google_play_scraper import reviews, Sort
import pandas as pd
import csv
```

```python
from google_play_scraper import reviews, Sort

webtoon_ripiu, _ = reviews(
    'com.naver.linewebtoon',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=15000
)
```

```python
ripiu_df = pd.DataFrame(webtoon_ripiu)

ripiu_clean_df = ripiu_df[['userName', 'content', 'score', 'at']]

ripiu_clean_df = ripiu_clean_df[ripiu_clean_df['content'].notnull()]
ripiu_clean_df = ripiu_clean_df[ripiu_clean_df['content'].str.len() > 10]

ripiu_clean_df.reset_index(drop=True, inplace=True)
ripiu_clean_df.to_csv('webtoon_ripiu.csv', index=False)

ripiu_df.head()
```
